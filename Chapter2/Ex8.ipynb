Full solution for the College data set exercise (parts a-h).

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Resolve the dataset path whether this notebook is run from Chapter2/ or repo root.
data_path = Path('College.csv')
if not data_path.exists():
    data_path = Path('Chapter2/College.csv')

print(f'Using data file: {data_path.resolve()}')
```

## (a) Read the data into Python as `college`

```python
college = pd.read_csv(data_path)
college.head()
```

## (b) Inspect and set college names as the index

```python
college2 = pd.read_csv(data_path, index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')

# Keep the modified version as requested.
college = college3
college.head()
```

## (c) Numerical summary with `describe()`

```python
college.describe(include='all').T
```

## (d) Scatterplot matrix of `[Top10perc, Apps, Enroll]`

```python
pd.plotting.scatter_matrix(
    college[['Top10perc', 'Apps', 'Enroll']],
    figsize=(9, 9),
    diagonal='hist',
    alpha=0.7
)
plt.suptitle('Scatterplot Matrix: Top10perc, Apps, Enroll', y=1.02)
plt.tight_layout()
plt.show()
```

## (e) Side-by-side boxplots of `Outstate` versus `Private`

```python
college.boxplot(column='Outstate', by='Private', figsize=(8, 5))
plt.title('Out-of-State Tuition by Private/Public')
plt.suptitle('')
plt.ylabel('Outstate')
plt.show()
```

## (f) Create `Elite` (Top10perc > 50), count, and boxplot

```python
college['Elite'] = pd.Categorical(np.where(college['Top10perc'] > 50, 'Yes', 'No'))

college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(8, 5))
plt.title('Out-of-State Tuition by Elite Status')
plt.suptitle('')
plt.ylabel('Outstate')
plt.show()
```

## (g) Histograms with different numbers of bins

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
cols = ['Apps', 'Accept', 'Enroll', 'Outstate']
bins = [10, 20, 30, 40]

for ax, col, b in zip(axes.ravel(), cols, bins):
    college[col].plot.hist(ax=ax, bins=b, edgecolor='black')
    ax.set_title(f'{col} (bins={b})')
    ax.set_xlabel(col)

plt.tight_layout()
plt.show()
```

## (h) Additional exploration and brief summary

```python
summary_private = college.groupby('Private')['Outstate'].median().rename('Median Outstate')
summary_elite = college.groupby('Elite')['Outstate'].median().rename('Median Outstate')
apps_corr = college.select_dtypes(include='number').corr(numeric_only=True)['Apps'].sort_values(ascending=False)

print('Median Outstate by Private/Public:')
print(summary_private)

print('Median Outstate by Elite status:')
print(summary_elite)

print('Top correlations with Apps:')
print(apps_corr.head(8))

print('Rows with Grad.Rate > 100 (possible data issue):')
print(college[college['Grad.Rate'] > 100][['Grad.Rate']])
```

### Brief findings
- Private schools generally charge higher out-of-state tuition than public schools.
- Elite schools (Top10perc > 50) are much more expensive out-of-state than non-elite schools.
- `Apps`, `Accept`, and `Enroll` are strongly positively related.
- Several quantitative variables are right-skewed (especially `Apps`, `Accept`, and `Enroll`).
- There is at least one questionable value (`Grad.Rate = 118`), which likely indicates a data-quality issue.
